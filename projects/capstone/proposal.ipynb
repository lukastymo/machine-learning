```python
from datetime import datetime

def toDate(ts):
    return datetime.fromtimestamp(ts).date()
```

```python
df = pd.read_csv("../../data/coinbaseUSD_1-min_data_2014-12-01_to_2018-06-27.csv")
# df["Date"] = datetime.fromtimestamp(df["Timestamp"]).isoformat()
df
```

```python
df["Date"] = df["Timestamp"].apply(lambda x: toDate(x))
idx = df.groupby("Date")["Timestamp"].idxmax()
df_max=df.loc[idx,]
df = df_max
df
```
